------------------------------------------------------------------------------

### Mean water depth

```python
# Read file
mean_wd_raster = rxr.open_rasterio(
    mean_wd_dir
)
```

```python
# Plot
mean_wd_raster.plot()
```

```python
# Remove nodata
mean_wd_raster_removenodata = mean_wd_raster.values[mean_wd_raster.values != -9999]
```

```python
plt.hist(mean_wd_raster_removenodata, bins=200);
```

```python
# 50th value
mean_wd_50th = np.quantile(mean_wd_raster_removenodata, 0.5)
mean_wd_50th
```

```python
# Get only below 50th data
mean_wd_lowerpart_50th = mean_wd_raster_removenodata[mean_wd_raster_removenodata <= mean_wd_50th]

# Get 30th value
mean_wd_lowerpart_50th_30th = np.quantile(mean_wd_lowerpart_50th, 0.3)

# The 30th value
mean_wd_lowerpart_50th_30th
```

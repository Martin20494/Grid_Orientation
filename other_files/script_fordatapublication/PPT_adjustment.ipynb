```python
# Mean
wse_mean = plt.imread(fr"{wse_dir}\\mean.png")
wd_mean = plt.imread(fr"{wd_dir}\\mean.png")

# Standard deviation
wse_sd = plt.imread(fr"{wse_dir}\\sd.png")
wd_sd = plt.imread(fr"{wd_dir}\\sd.png")

# Coefficient of variation
wd_cv = plt.imread(fr"{wd_dir}\\cv.png")

# Proportion
wd_cell = plt.imread(fr"{wd_dir}\\cell.png")

# Building
building = plt.imread(fr"{wd_dir}\\building_nodensity.png")

# Area
area = plt.imread(fr"{wd_dir}\\area_nodensity.png")

# Problem
problem = plt.imread(fr"{problem_dir}\\problem_idea.png")

# Transformation process
trans_process = plt.imread(fr"{trans_process}\\transformation_process.png")
```

#### 2. Get margins

```python
# Ref: https://stackoverflow.com/questions/6541123/improve-subplot-size-spacing-with-many-subplots
# https://stackoverflow.com/questions/41210823/using-plt-imshow-to-display-multiple-images

left  = 0.1  # the left side of the subplots of the figure
right = 0.9   # the right side of the subplots of the figure
bottom = 0.1   # the bottom of the subplots of the figure
top = 0.9      # the top of the subplots of the figure
wspace = 0.03  # the amount of width reserved for blank space between subplots
hspace = 0.001   # the amount of height reserved for white space between subplots
```

#### 3. Results

```python
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(17, 30))

plt.subplots_adjust(left=left, bottom=bottom, right=right, top=top, wspace=wspace, hspace=hspace)

ax[0].imshow(wd_mean)
ax[0].axis('off')
ax[1].imshow(wse_mean)
ax[1].axis('off')

plt.savefig(fr"{save_dir}\mean_combination.png", bbox_inches='tight', dpi=500)
```

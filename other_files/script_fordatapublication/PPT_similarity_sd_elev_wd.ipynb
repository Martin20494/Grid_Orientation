```python
# Packages
import matplotlib.pyplot as plt
import os
```

```python
# Get path
main_path = r"P:\Martin_publication\Grid_Orientation\Ppt_adjustment"
similarity_path = fr"{main_path}\\similarity_sd_elev_wd.jpg"
```

```python
similarity_plot = plt.imread(similarity_path)
```

```python
left  = 0.1  # the left side of the subplots of the figure
right = 0.9   # the right side of the subplots of the figure
bottom = 0.1   # the bottom of the subplots of the figure
top = 0.9      # the top of the subplots of the figure
wspace = 0.03  # the amount of width reserved for blank space between subplots
hspace = 0.001   # the amount of height reserved for white space between subplots
```

```python
fig, ax = plt.subplots(figsize=(17, 30))

plt.subplots_adjust(left=left, bottom=bottom, right=right, top=top, wspace=wspace, hspace=hspace)

ax.imshow(similarity_plot)
ax.axis('off')

plt.savefig(fr"{main_path}\S3_similarity_sd_elev_wd.jpg", bbox_inches='tight', dpi=350)
```

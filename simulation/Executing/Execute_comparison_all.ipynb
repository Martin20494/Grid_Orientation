```python
import matplotlib
import colorsys
import matplotlib.colors as mc
```

```python
def lighten_color(color, amount=0.5):
    """
    @Definition:
                A function to lighten the color (by @IanHincks)
    @References:
                https://stackoverflow.com/questions/37765197/darken-or-lighten-a-color-in-matplotlib/49601444#49601444
                https://stackoverflow.com/questions/55656683/change-seaborn-boxplot-line-rainbow-color
    @Arguments:
                color (matplotlib color):
                                A tuple of 3 values of RGB color
                amount (float):
                                Level of brightness of color
    @Returns:
                A new tuple of 3 values of RGB color
    """
    try:
        c = mc.cnames[color]
    except:
        c = color
    c = colorsys.rgb_to_hls(*mc.to_rgb(c))
    return colorsys.hls_to_rgb(c[0], 1 - amount * (1 - c[1]), c[2])
```

```python
# Set up ax[p]is
fig, ax = plt.subplots(3, 2, figsize=(25, 40))

# Filter -9999
plt.subplots_adjust(wspace=.3, hspace=.55)

x_label = ['Coefficient of variation (%)', 'Number of buildings']

data_df_list = [
    [trans_comparison['cv'], trans_comparison['building']],
    [res_comparison['cv'], res_comparison['building']],
    [events_comparison['cv'], events_comparison['building']]
]

# Fontsize
fontsize = 24
labelpad = 29

for row in range(3):
    for col in range(2):
        # Boxplot
        boxplot_sns = sns.boxplot(
            data=data_df_list[row][col],
            orient='h', # Boxplots lie horizontally
            showmeans=True, # Turn on mean sign
            meanprops=dict(marker='s', markersize=6), # Use big dot to visualise mean sign
            flierprops=dict(marker='o', markersize=1.5), # Visualise outliers
            width=0.4, # Size/width of boxplots
            palette='husl', saturation=1, ax=ax[row][col]
        )
        
        # Legend
        if col == 0 and row ==0:
            ax[row][col].plot([], [], '|', linewidth=1, color='black', label='median')
            ax[row][col].plot([], [], 's', linewidth=1, color='black', label='mean')
            ax[row][col].legend(fontsize=fontsize, frameon=False, loc='upper right', markerscale=2)
        else:
            pass

        # Colorise all lines of boxplots
        box_patches = [
            patch for patch in ax[row][col].patches if type(patch) == matplotlib.patches.PathPatch
        ]
        num_patches = len(box_patches)
        lines_per_boxplot = len(ax[row][col].lines) // num_patches
        for i, patch in enumerate(box_patches):
            # Set the linecolor on the patch to the facecolor, and set the facecolor to None
            colu = lighten_color(patch.get_facecolor(), 1.5)
            patch.set_edgecolor(colu)

            # Each box has associated Line2D objects (to make the whiskers, fliers, etc.)
            # Loop over them here, and use the same color as above
            for line in ax[row][col].lines[i * lines_per_boxplot: (i + 1) * lines_per_boxplot]:
                line.set_color(colu)
                line.set_mfc(colu)  # facecolor of fliers
                line.set_mec(colu)  # edgecolor of fliers
                line.set_linewidth(0.7)

        # Adjust x and y labels
        ax[row][col].set_xlabel(x_label[col], fontsize=fontsize, labelpad=labelpad)
        ax[row][col].set(ylabel=None)
        if col!=0:
            boxplot_sns.set(yticklabels=[])
            boxplot_sns.tick_params(left=False)
        else:
            pass

        # Design size and style for ticks and labels
        for item in (ax[row][col].get_xticklabels() + ax[row][col].get_yticklabels()):  # For x, y ticks' labels
            item.set_fontsize(fontsize-3)

        ax[row][col].tick_params(direction='out', length=5, pad=labelpad-17)
        

# TRANS - CV data
trans_col1 = 'Quartile'
trans_col2 = 'deviation'
trans_cv_results = comparison_calculation(data_df_list[0][0], 'quartile dev')
trans_cv_text_full = "".join("{:.3f}\n".format(trans_cv_results[i]) for i in range(len(trans_cv_results)))
trans_cv_text_rm = trans_cv_text_full.rstrip()
ax[0][0].text(660, -1.2, trans_col1, fontsize=fontsize, horizontalalignment='center', weight='bold',
                  verticalalignment='center', linespacing=3.8)
ax[0][0].text(660, -.8, trans_col2, fontsize=fontsize, horizontalalignment='center', weight='bold',
                  verticalalignment='center', linespacing=3.8)
ax[0][0].text(660, 2, trans_cv_text_rm, fontsize=fontsize, horizontalalignment='center',
                  verticalalignment='center', linespacing=5.6)
# TRANS - BUILDING data
trans_col1 = 'Quartile'
trans_col2 = 'deviation'
trans_building_results = comparison_calculation(data_df_list[0][1], 'quartile dev')
trans_building_text_full = "".join("{:.3f}\n".format(trans_building_results[i]) for i in range(len(trans_building_results)))
trans_building_text_rm = trans_building_text_full.rstrip()
ax[0][1].text(215, -1.2, trans_col1, fontsize=fontsize, horizontalalignment='center', weight='bold',
                  verticalalignment='center', linespacing=3.8)
ax[0][1].text(215, -.8, trans_col2, fontsize=fontsize, horizontalalignment='center', weight='bold',
                  verticalalignment='center', linespacing=3.8)
ax[0][1].text(215, 2, trans_building_text_rm, fontsize=fontsize, horizontalalignment='center',
                  verticalalignment='center', linespacing=5.6)
# TRANS - (a)
font = {'fontname':'Times New Roman'}
ax[0][0].text(660, 5.6, '(a)', fontsize=fontsize+17, horizontalalignment='center',
                  verticalalignment='center', linespacing=5, **font)
# ---------------------------------------------------------------




# RES - CV data
res_col1 = 'Quartile'
res_col2 = 'deviation'
res_cv_results = comparison_calculation(data_df_list[1][0], 'quartile dev')
res_cv_text_full = "".join("{:.3f}\n".format(res_cv_results[i]) for i in range(len(res_cv_results)))
res_cv_text_rm = res_cv_text_full.rstrip()
ax[1][0].text(608, -.95, res_col1, fontsize=fontsize, horizontalalignment='center', weight='bold',
                  verticalalignment='center', linespacing=3.8)
ax[1][0].text(608, -.7, res_col2, fontsize=fontsize, horizontalalignment='center', weight='bold',
                  verticalalignment='center', linespacing=3.8)
ax[1][0].text(608, 1.48, res_cv_text_rm, fontsize=fontsize, horizontalalignment='center',
                  verticalalignment='center', linespacing=7)
# RES - BUILDING data
res_col1 = 'Quartile'
res_col2 = 'deviation'
res_building_results = comparison_calculation(data_df_list[1][1], 'quartile dev')
res_building_text_full = "".join("{:.3f}\n".format(res_building_results[i]) for i in range(len(res_building_results)))
res_building_text_rm = res_building_text_full.rstrip()
ax[1][1].text(630, -.95, res_col1, fontsize=fontsize, horizontalalignment='center', weight='bold',
                  verticalalignment='center', linespacing=3.8)
ax[1][1].text(630, -.7, res_col2, fontsize=fontsize, horizontalalignment='center', weight='bold',
                  verticalalignment='center', linespacing=3.8)
ax[1][1].text(630, 1.48, res_building_text_rm, fontsize=fontsize, horizontalalignment='center',
                  verticalalignment='center', linespacing=7)

# RES - (b)
font = {'fontname':'Times New Roman'}
ax[1][0].text(608, 4.45, '(b)', fontsize=fontsize+17, horizontalalignment='center',
                  verticalalignment='center', linespacing=9.5, **font)
# ---------------------------------------------------------------


# EVENT - CV data
event_col1 = 'Quartile'
event_col2 = 'deviation'
event_cv_results = comparison_calculation(data_df_list[2][0], 'quartile dev')
event_cv_text_full = "".join("{:.3f}\n".format(event_cv_results[i]) for i in range(len(event_cv_results)))
event_cv_text_rm = event_cv_text_full.rstrip()
ax[2][0].text(497, -1.4, event_col1, fontsize=fontsize, horizontalalignment='center', weight='bold',
                  verticalalignment='center', linespacing=3.8)
ax[2][0].text(497, -.9, event_col2, fontsize=fontsize, horizontalalignment='center', weight='bold',
                  verticalalignment='center', linespacing=3.8)
ax[2][0].text(497, 2.5, event_cv_text_rm, fontsize=fontsize, horizontalalignment='center',
                  verticalalignment='center', linespacing=4.5)
# EVENT - BUILDING data
event_col1 = 'Quartile'
event_col2 = 'deviation'
event_building_results = comparison_calculation(data_df_list[2][1], 'quartile dev')
event_building_text_full = "".join("{:.3f}\n".format(event_building_results[i]) for i in range(len(event_building_results)))
event_building_text_rm = event_building_text_full.rstrip()
ax[2][1].text(285, -1.4, event_col1, fontsize=fontsize, horizontalalignment='center', weight='bold',
                  verticalalignment='center', linespacing=3.8)
ax[2][1].text(285, -.9, event_col2, fontsize=fontsize, horizontalalignment='center', weight='bold',
                  verticalalignment='center', linespacing=3.8)
ax[2][1].text(285, 2.5, event_building_text_rm, fontsize=fontsize, horizontalalignment='center',
                  verticalalignment='center', linespacing=4.5)

# EVENT - (c)
font = {'fontname':'Times New Roman'}
ax[2][0].text(497, 6.9, '(c)', fontsize=fontsize+17, horizontalalignment='center',
                  verticalalignment='center', linespacing=4.5, **font)


# # Building data
# else:
#     col1 = 'Quantile'
#     col2 = 'deviation'
#     building_results = comparison_calculation(data_df_list[row][col], 'quartile dev')
#     building_text_full = "".join("{:.3f}\n".format(building_results[i]) for i in range(len(building_results)))
#     building_text_rm = building_text_full.rstrip()
#     ax[row][col].text(1, -.95, col1, fontsize=fontsize, horizontalalignment='center', weight='bold',
#                       verticalalignment='center', linespacing=3.8)
#     ax[row][col].text(1, -.7, col2, fontsize=fontsize, horizontalalignment='center', weight='bold',
#                       verticalalignment='center', linespacing=3.8)
#     ax[row][col].text(1, 1, building_text_rm, fontsize=fontsize, horizontalalignment='center',
#                       verticalalignment='center', linespacing=7.6)

# Save fig
plt.savefig(
    fr"S:\\AGU\\data\\comparison_all_003.png",
    bbox_inches='tight', dpi=330
)
```
